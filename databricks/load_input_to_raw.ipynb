###load_input_to_raw
####Reads raw CSV files from ADLS landing zone and loads data into the Bronze layer (ecom_raw) Delta tables with ingestion metadata

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
dbutils.widgets.text("load_type", "")
dbutils.widgets.text("pipeline_run_id","")
dbutils.widgets.text("ingestion_date","")
dbutils.widgets.text("processing_year","")
```

```python
load_type = dbutils.widgets.get("load_type")
pipeline_run_id = dbutils.widgets.get("pipeline_run_id")
ingestion_date = dbutils.widgets.get("ingestion_date")
processing_year = dbutils.widgets.get("processing_year")

if load_type == "static":
  load = "ref"
else:
  load = "config"
```

```python
%run /Workspace/e-commerce/Parameter_setup
```

####Customers

```python
customers_df = spark.read.csv(f"{inbound_path}/{processing_year}_customers_dataset.csv", header=True, inferSchema=True)
customers_df = (customers_df.withColumn("ingestion_date", to_timestamp(lit(ingestion_date)))
                            .withColumn("pipeline_run_id",lit(pipeline_run_id))
                )
customers_df.createOrReplaceTempView("customers")
```

```python
spark.sql(f'''
INSERT INTO ecom_raw.customers 
SELECT * FROM customers
''').display()
```
